### 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

#### Окружность

```python
x = np.linspace(-10, 10, 100)
R = 10
y = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
t = np.arange(0, np.pi * 2.0, 0.01)
r = 10
x = r * np.cos(t)
y = r * np.sin(t)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

#### Эллипс

```python
x = np.linspace(-10, 10, 150)
a = 10
b = 3
y = np.sqrt((1-x**2/a**2)*b**2)
y2 = -np.sqrt((1-x**2/a**2)*b**2)

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
t = np.arange(0, np.pi * 2.0, 0.01)
a = 10
b = 3
x = a * np.cos(t)
y = b * np.sin(t)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

#### Гипербола

```python
x = np.linspace(-20, 20, 150)
a = 3
b = 2
y = np.sqrt((1+x**2/a**2)*b**2)
y2 = -np.sqrt((1+x**2/a**2)*b**2)

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
X,Y = np.meshgrid(x,y)
Z= 4*X + 8*Y + 1

X1,Y1 = np.meshgrid(x,y)
Z1= 4*X1 + 8*Y1 + 10

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X1, Y1, Z1)
plt.show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.
