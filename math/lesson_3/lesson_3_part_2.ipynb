### 1. Задание (в программе)
Нарисуйте график функции:

y(x) = k∙cos(x – a) + b

для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)

k = 2
a = 5
b = 3

y = k * np.cos(x - a) + b
plt.plot(x, y, label=f'$y = {k}∙cos(x - {a}) + {b}$')

k = 1
a = 9
b = 1

y = k * np.cos(x - a) + b
plt.plot(x, y, label=f'$y = {k}∙cos(x - {a}) + {b}$')

k = 2
a = 1
b = 2

y = k * np.cos(x - a) + b
plt.plot(x, y, label=f'$y = {k}∙cos(x - {a}) + {b}$')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.

Напишите код, который будет рисовать график окружности в полярных координатах.

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
t = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)

plt.polar(x, y)
plt.show()
```

Решите систему уравнений:
$\left\{

y = x2 – 1 \\

\right.$

```python
x = np.linspace(-6, 6, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-3, 5, 200)

y = np.exp(x) - 1 + x*(2-x**2)
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equations(x):
    return np.exp(x) - 1 + x*(2-x**2)
```

```python
print(f'x_1:{fsolve(equations, -1.5)[0]}\t y_1: {(fsolve(equations, -1.5)[0])**2 - 1}\n',\
      f'x_2:{fsolve(equations, 0)[0]}\t\t y_2: {(fsolve(equations, 0)[0])**2 - 1}\n',\
      f'x_3:{fsolve(equations, 2.5)[0]}\t y_3: {(fsolve(equations, 2.5)[0])**2 - 1}\n',\
      f'x_4:{fsolve(equations, 4)[0]}\t y_4: {(fsolve(equations, 4)[0])**2 - 1}'
     )
```
